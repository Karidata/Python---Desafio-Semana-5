```python
# Parte 1: Incidentes, oficiales y sujeto
```

```python
# 1.a Carga de datos
```

```python
import pandas as pd

# Cargamos los archivos pickle
incidents = pd.read_pickle("incidents.pkl")
officers = pd.read_pickle("officers.pkl")
subjects = pd.read_pickle("subjects.pkl")

# Verificamos las primeras filas
incidents.head(), officers.head(), subjects.head()
```

```python
# 1.b Unión de las tres tablas
```

```python
# Unimos incidents con officers
df_merge1 = pd.merge(incidents, officers, on="case_number", suffixes=("_incident", "_officer"))

# Unimos el resultado con subjects
df_final = pd.merge(df_merge1, subjects, on="case_number", suffixes=("", "_subject"))

df_final.head()
```

```python
# 1.c Eliminación de duplicados
```

```python
# Verificamos duplicados
duplicados = df_final.duplicated().sum()

# Eliminamos duplicados si existen
df_final = df_final.drop_duplicates()

duplicados
```

```python
# 1.d Conteo de sujetos de género femenino
```

```python
# Contamos cuántos sujetos tienen género femenino ('F') en la columna 'gender'
conteo_genero = df_final["gender"].value_counts()

# Mostramos el resultado de forma clara
print("Número de sujetos de género femenino (F):", conteo_genero.get("F", 0))
```

```python
# 1.e Número de casos con al menos una mujer sospechosa
```

```python
# Filtramos por género femenino
casos_mujer = df_final[df_final["gender"] == "F"]["case_number"].unique()
len(casos_mujer)
```

```python
# 1.f Tabla pivote por género de oficial vs sujeto
```

```python
# Creamos tabla pivote entre género del oficial y género del sujeto
pivot_genero = pd.pivot_table(df_final, index="gender", columns="gender_subject", aggfunc="size", fill_value=0)

# Mostramos la tabla
print(pivot_genero)
```

```python
import numpy as np
import pandas as pd

# Función segura para extraer salario mínimo y máximo
def extraer_salarios(s):
    try:
        s = str(s).replace("$", "").replace("K", "").replace("per year", "").strip()
        min_sal, max_sal = s.split("-")
        return float(min_sal), float(max_sal)
    except:
        return np.nan, np.nan

# Aplicamos la función y convertimos explícitamente a float para evitar conflictos de tipo
salarios = df_jobs_clean["Salary Estimate"].apply(lambda x: pd.Series(extraer_salarios(x))).astype(float)

# Asignamos las columnas de forma segura
df_jobs_clean.loc[:, "Salario Estimado Mínimo"] = salarios[0]
df_jobs_clean.loc[:, "Salario Estimado Máximo"] = salarios[1]

# Generamos la tabla pivote
pivot_salario = pd.pivot_table(
    df_jobs_clean,
    index="Size",
    values=["Salario Estimado Mínimo", "Salario Estimado Máximo"],
    aggfunc="mean"
)

# Mostramos la tabla pivote
print(pivot_salario)
```
